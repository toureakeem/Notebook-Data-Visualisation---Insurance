## Insurance

This program is to investigate the relationship between age and insurance and plot a graph based on the results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# First we import the data from salary.csv and view how the data looks
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = insurance_data.iloc[:,:1].values
y = insurance_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

This part of the code is to create the line of best fit.

```python
# Fit the model
insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

Below would be the code to add the line of best fit with the help of linear regression

```python
# Plot the data and model
y_pred = insurance_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
